```python
import os
import cv2


print(f"Current working directory: {os.getcwd()}")

# change for your own directories and files names
masks_directory = "./isles/masks/"
labels_directory = "./isles/labels/"


os.makedirs(labels_directory, exist_ok=True)

def convert_mask_to_yolo_format(mask_file, masks_dir, labels_dir):

    mask_path = os.path.join(masks_dir, mask_file)
    
    
    mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)

   
    if mask is None:
        print(f"Error: Could not read image {mask_path}")
        return
    
   
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

  
    label_file = os.path.join(labels_dir, mask_file.replace(".png", ".txt"))
    
  
    
   
    with open(label_file, "w") as file:
        for contour in contours:
        
            x, y, w, h = cv2.boundingRect(contour)
            
       
            mask_height, mask_width = mask.shape

            # Normalize the coordinates relative to the image size
            x_center = (x + w / 2) / mask_width
            y_center = (y + h / 2) / mask_height
            width = w / mask_width
            height = h / mask_height

            # Class ID is always 0 (since you're looking for the tumor)
            file.write(f"0 {x_center} {y_center} {width} {height}\n")

    print(f"Converted {mask_file} to YOLO format.")


mask_files = [f for f in os.listdir(masks_directory) if f.endswith(".png")]


for mask_file in mask_files:
    convert_mask_to_yolo_format(mask_file, masks_directory, labels_directory)
```
